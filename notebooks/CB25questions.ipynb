Question 2 Show all cricket matches that were played in the last Few days. Include the match description, both team names, venue name with city, and the match date. Sort by most recent matches first.

```python
import requests

url = "https://cricbuzz-cricket.p.rapidapi.com/matches/v1/recent"

headers = {
	"x-rapidapi-key": "f28eedda78mshf250801249f0213p108557jsn01ea9788bf50",
	"x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com",
	"Content-Type": "application/json"
}

response = requests.get(url, headers=headers)
RM_rawData=(response.json())

RM_rawData['typeMatches'][0]['seriesMatches'][0]['seriesAdWrapper']['matches'][0]['matchInfo']

RM_matchInfo=[]
for type_match in RM_rawData['typeMatches']:
    for series_match in type_match['seriesMatches']:
        wrapper = series_match.get('seriesAdWrapper')
        if not wrapper:
            continue
        for match in wrapper.get('matches', []):
            RM_matchInfo.append(match['matchInfo'])
            print(match['matchInfo'])
```

```python
#recentMatches_info_table

RM_rows = []

for info in RM_matchInfo:

    start_ts = info.get('startDate')
    end_ts = info.get('endDate')

   #start_date = datetime.fromtimestamp(int(start_ts) / 1000).strftime('%d/%m/%Y') if start_ts else None
    #end_date = datetime.fromtimestamp(int(end_ts) / 1000).strftime('%d/%m/%Y') if end_ts else None

    start_date = datetime.fromtimestamp(int(start_ts) / 1000) if start_ts else None
    end_date = datetime.fromtimestamp(int(end_ts) / 1000) if end_ts else None

    row = {
        "matchId": info.get('matchId'),
        "seriesId": info.get('seriesId'),
        "seriesName": info.get('seriesName'),
        "matchDesc": info.get('matchDesc'),
        "startDate": start_date,
        "endDate": end_date,
        "team1": info['team1']['teamSName'],
        "team2": info['team2']['teamSName'],
        "venue" : info['venueInfo']['ground'],
        "city": info['venueInfo']['city'],
        "status": info.get('status')
    }

    RM_rows.append(row)

RM = pd.DataFrame(RM_rows)
print(RM)
```

Question 3 List the top 10 highest run scorers in ODI cricket. Show player name, total runs scored, batting average, and number of centuries. Display the highest run scorer first.

```python
import requests
import pandas as pd

url = "https://cricbuzz-cricket.p.rapidapi.com/stats/v1/topstats/0"

headers = {
    "x-rapidapi-key": "f28eedda78mshf250801249f0213p108557jsn01ea9788bf50",
    "x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

# --- Fetch Most Runs (ODI) ---
response1 = requests.get(url, headers=headers, params={"statsType": "mostRuns", "matchTypeId": "2"})
runs_data = response1.json().get("values", [])

# --- Fetch Most Hundreds (ODI) ---
response2 = requests.get(url, headers=headers, params={"statsType": "mostHundreds", "matchTypeId": "2"})
hundreds_data = response2.json().get("values", [])

# --- Build Lookup: Player ID -> Hundreds ---
hundreds_by_id = {p["values"][0]: p["values"][5] for p in hundreds_data}

# --- Sort by Runs Descending ---
runs_sorted = sorted(runs_data, key=lambda x: int(x["values"][4]), reverse=True)

# --- Build DataFrame ---
records = []
for player in runs_sorted[:10]:
    player_id = player["values"][0]
    records.append({
        "Player ID" : player_id,
        "Player"    : player["values"][1],
        "Runs"      : int(player["values"][4]),
        "Average"   : float(player["values"][5]),
        "100s"      : int(hundreds_by_id.get(player_id, 0))
    })

most_run_scores = pd.DataFrame(records)
most_run_scores.index = range(1, len(most_run_scores) + 1)  # Rank starting from 1
most_run_scores.index.name = "Rank"

print(most_run_scores.to_string())
```

Question 4 Display all cricket venues that have a seating capacity of more than 25,000 spectators. Show venue name, city, country, and capacity. Order by largest capacity first (10 Venues enough).

```python
import time
import requests
import pandas as pd

headers = {
    "x-rapidapi-key": "f28eedda78mshf250801249f0213p108557jsn01ea9788bf50",
    "x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com",
    "Content-Type": "application/json"
}

valid_data = []
error_data = []

for venue_id in range(1, 500):
    url = f"https://cricbuzz-cricket.p.rapidapi.com/venues/v1/{venue_id}"

    try:
        response = requests.get(url, headers=headers, timeout=10)

        if response.status_code != 200:
            error_data.append({
                "id": venue_id,
                "ground": "N/A",
                "capacity": 0,
                "city": "N/A",
                "country": "N/A",
                "error": f"HTTP {response.status_code}"
            })
            continue

        data = response.json()

        if not data:
            error_data.append({
                "id": venue_id,
                "ground": "N/A",
                "capacity": 0,
                "city": "N/A",
                "country": "N/A",
                "error": "Empty JSON"
            })
            continue

        ground = data.get("ground", "N/A")
        capacity = data.get("capacity", "0")
        city = data.get("city", "N/A")
        country = data.get("country", "N/A")

        try:
            capacity = int(str(capacity).replace(",", ""))
        except:
            capacity = 0

        valid_data.append({
            "id": venue_id,
            "ground": ground,
            "capacity": capacity,
            "city": city,
            "country": country
        })

    except Exception as e:
        error_data.append({
            "id": venue_id,
            "ground": "N/A",
            "capacity": 0,
            "city": "N/A",
            "country": "N/A",
            "error": str(e)
        })

    time.sleep(3)

# --- CONVERT TO DATAFRAMES ---
df_valid = pd.DataFrame(valid_data)

# --- TOP 10 STADIUMS BY CAPACITY ---
venue_top10 = df_valid.sort_values(by="capacity", ascending=False).head(10).reset_index(drop=True)
venue_top10.index += 1
venue_top10.index.name = "Rank"

# --- OUTPUT ---
print("\n🔥 TOP 10 STADIUMS (BY CAPACITY):")
print(venue_top10.to_string())

venue_list25k = df_valid[df_valid["capacity"] >= 25000].sort_values(by="capacity", ascending=False)
print(venue_list25k.to_string(index=False))
```

```python
import os
from dotenv import load_dotenv
import pandas as pd
from sqlalchemy import create_engine

# Load local .env.cricbuzz for local development
load_dotenv(dotenv_path=Path("../.env.cricbuzz"), override=True)

# Get MySQL credentials from environment
user = os.getenv("MYSQL_USER")
password = os.getenv("MYSQL_PWD")
host = os.getenv("MYSQL_HOST")
port = os.getenv("MYSQL_PORT")
database = os.getenv("MYSQL_DB")


engine = create_engine(
    f"mysql+mysqlconnector://{user}:{password}@{host}:{port}/{database}")
venue_list25k.to_sql(name="bigger_venues", con=engine, if_exists="replace", index=False)

print(f"Data uploaded to MySQL successfully, Database: {database}, Table: bigger_venues")
```

Question 5 Calculate how many matches each team has won. Show team name and total number of wins. Display teams with the most wins first.

```python
import requests
import pandas as pd

url = "https://cricbuzz-cricket.p.rapidapi.com/stats/v1/series/9237/points-table"

headers = {
	"x-rapidapi-key": "f28eedda78mshf250801249f0213p108557jsn01ea9788bf50",
	"x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com",
	"Content-Type": "application/json"
}

response = requests.get(url, headers=headers)
data = response.json()

teams = data["pointsTable"][0]["pointsTableInfo"]

df = pd.DataFrame([
    {
        "team_name": t["teamFullName"],
        "total_wins": t["matchesWon"]
    }
    for t in teams
]).sort_values("total_wins", ascending=False).reset_index(drop=True)

df.index += 1
df.index.name="Sl.no"
print(df.to_string())
```
